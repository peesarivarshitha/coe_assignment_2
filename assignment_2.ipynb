# Part 2-City-Wise Analysis

```python
#find total sales for each city in each month using groupby()
total_sales = sales.groupby("city")[["Jan", "Feb", "Mar", "Apr"]].sum()
print(total_sales)
```

```python
#find the average monthly sales per city
sales['Average_Monthly_Sales'] = sales[['Jan', 'Feb', 'Mar', 'Apr']].mean(axis=1)
print(sales[['city', 'Average_Monthly_Sales']])
```

```python
#the city with the highest average sales
avg_city = sales.loc[sales['Average_Monthly_Sales'].idxmax(), ['city', 'Average_Monthly_Sales']]
avg_city
```

```python
#the city with the lowest total sales
sales['Total_Sales'] = sales[['Jan', 'Feb', 'Mar', 'Apr']].sum(axis=1)
total_city = sales.loc[sales['Total_Sales'].idxmin(), ['city', 'Total_Sales']]
total_city
```

```python
#add a new column total_sales=jan+feb+mar+apr
sales['Total_Sales']=sales[['Jan','Feb','Mar','Apr']].sum(axis=1)
sales.head()
```

```python
#find the top 3 performing cities based on total Sales
top_cities = sales.sort_values(by='Total_Sales', ascending=False).head(3)
top_cities
```

# Part 3-Store-level insights

```python
#display the top 5 stores with the highest total_sales
highest=sales.sort_values(by='Total_Sales',ascending=False).head(5)
highest[['store_id','Total_Sales']]
```

```python
##display the bottom 5 stores with the lowest total_sales
lowest=sales.sort_values(by='Total_Sales',ascending=True).head(5)
lowest[['store_id','Total_Sales']]
```

```python
#the store with maximum sales in March
sales.loc[sales['Mar'].idxmax(), ['store_id', 'Mar']]
```

```python
#find how many stores in each city have average monthly sales>15
sales['avg_sales'] = sales.loc[:, ['Jan', 'Feb', 'Mar', 'Apr']].mean(axis=1)
avg_sales = sales[sales['avg_sales'] > 15]
stores_per_city=avg_sales.groupby('city')['store_id'].count()
stores_per_city

```

```python
# Filter stores from California that had sales > 20 in any month.
#sales[(sales['city']=='California')&((sales['Jan']>20)|(sales['Feb']>20)|(sales['Mar']>20)|(sales['Apr']>20))][['store_id','Jan','Feb','Mar','Apr','Total_Sales']]
california_stores=sales[sales['city'] == 'California']
california_high_sales = california_stores[(california_stores.loc[:, ['Jan', 'Feb', 'Mar', 'Apr']] > 20).any(axis=1)]
california_high_sales
```

# Part 4-Aggregation and Analysis

```python
#mean,max,and min for each month
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
          'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales[months].agg(['mean', 'max', 'min'])

```

```python
#Find the city with the highest average march sales
march_avg = sales.groupby('city')['Mar'].mean()
top_city = march_avg.idxmax()
top_sales = march_avg.max()
top_city, top_sales
```

```python
#calculate the overall average sales per city(across all months)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales['avg_sales'] = sales[months].mean(axis=1)
city_avg_sales = sales.groupby('city')['avg_sales'].mean()
city_avg_sales
```

```python
#identify the month with the highest overall sales
top_month = sales.sum(axis=0)[['Jan','Feb','Mar','Apr','May','Jun','July','Aug','Sep','Oct','Nov','Dec']].idxmax()
top_month
```

```python
#for each city,find the month with maximum total sales
city_total_sales = sales.groupby('city')[months].sum()
max_total_sales= city_total_sales.idxmax(axis=1)
max_total_sales
```

```python
#add a column average_sales(average of all months per store
sales['Average_Sales'] = sales[months].mean(axis=1)
sales.head()
```

```python
#sort the dataset by this column in descending order
sales.sort_values(by='city', ascending=False)
```
